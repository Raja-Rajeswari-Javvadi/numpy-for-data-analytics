This notebook demonstrates how to work with external CSV files using NumPy. It includes reading and writing data, performing analysis on file-based datasets, and understanding the limitations of NumPy, thereby creating a smooth transition toward using Pandas for advanced data analytics.


```python
import numpy as np
```

```python
#Create Sample CSV
data = np.array([[101,5000,2], # employee id , salary , experience
                 [102,5500,7],
                 [103,6000,3],
                 [104,7700,6],
                 [105,5330,1]])
np.savetxt("employees.csv",data,delimiter=",",fmt="%d")
```

```python
# Read CSV Using NumPy
emp_data = np.loadtxt("employees.csv",delimiter=",",dtype=np.int64)
print(emp_data)
```

```python
# Analyze File Data
# Average salary
avg_salary = np.mean(emp_data[:,1])
print(avg_salary)
```

```python
# Highest experience
high_exp = np.max(emp_data[:,2])
print(high_exp)
```

```python
# Employees with salary > 5500
high_salary_employees = data[data[:, 1] > 5500]
high_salary_employees
```
